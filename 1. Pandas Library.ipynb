```python
toyo_df=pd.read_csv('Toyota.csv',index_col=0)
```

```python
toyo_df.head()
```

```python
np.unique(toyo_df['Doors'])
```

```python
# in doors it is mix of int & str

toyo_df['Doors'].replace('three',3,inplace=True)
toyo_df['Doors'].replace('four',4,inplace=True)
toyo_df['Doors'].replace('five',5,inplace=True)
```

```python
np.unique(toyo_df['Doors'])
```

```python
toyo_df.info()
```

```python
### assigning the type to data
toyo_df['Doors']=toyo_df['Doors'].astype('int64')
```

```python
np.unique(toyo_df['Doors'])
```

```python
#added new column
toyo_df.insert(10,'Price_class',' ')
```

```python
toyo_df.head()
```

```python
toyo_df.head()
```

```python
# price class - classification as  - add categoty
# Low -> Upto 8450
# Medium -> 8450 - 11950
# High -> > 11950
```

```python
# new list - not inplace
Price_class = []
for i in range(len(toyo_df))

    if toyo_df["Price"][i]<=8450:
        Price_class.append('Low')
    elif toyo_df["Price"][i]<=11950:
        Price_class.append('Medium')
    else:
        Price_class.append('High')
```

```python
# inplace 

for i in range(len(toyo_df['Price'])):
    if toyo_df["Price"][i]<=8450:
        toyo_df['Price_class'][i]='Low'
    elif toyo_df["Price"][i]<=11950:
        toyo_df['Price_class'][i]='Medium'
    else:
        toyo_df['Price_class'][i]='High'
```
